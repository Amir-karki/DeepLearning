```python
import pandas as pd
df = pd.read_csv("./Dataset/spamorham.csv")
df
```

```python
# Data cleaning and preprocessing
import re
import nltk
nltk.download("stopwords")
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['sms'][i].lower()).split()  # substitute regular exp. with {tab} and lowercase all words then apply split
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')] # return only those words that are not present in stopwords.
    review = ' '.join(review) 
    corpus.append(review)
```

```python
corpus
```

#### create the bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=500, binary=True)
bow = cv.fit_transform(corpus).toarray()
bow
```

```python
bow.shape
```

```python
cv.vocabulary_ # returns what vocabularies are included in the bow
```

```python
cv2 = CountVectorizer(max_features=1500, binary=True, ngram_range=(1,2))
bow2 = cv2.fit_transform(corpus).toarray()
bow2
```
